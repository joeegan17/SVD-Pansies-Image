```python
import numpy as np
import matplotlib.image as img
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from PIL import Image
```

```python
svd = np.linalg.svd(image)
```

```python
print(svd[0].shape)
print(svd[1].shape)
print(svd[2].shape)
```

```python
u = svd[0]
diag = svd[1]
v_transpose = svd[2]
v = svd[2].transpose()
```

```python
diag[0:10]
```

```python
np.linalg.matrix_rank(image) == len(diag)
#rank of matrix is equal to # of non-zero singular values
```

```python
singular_vals = pd.DataFrame({
    'Val_Number' : np.arange(0,1225), 'Singular_Values' : diag
})
```

```python
sns.lineplot(x=singular_vals.Val_Number, y=singular_vals.Singular_Values )
plt.title('Singular Values in Descending Order by Magnitude')
plt.show;
```

```python
img_10 = np.matrix(u[:,:10]) * np.diag(diag[:10])*np.matrix(v_transpose[:10, :])
plt.figure(figsize=(20,10))
plt.imshow(img_10, cmap='Greys', interpolation='None')
plt.axis('off')
plt.show;
```
